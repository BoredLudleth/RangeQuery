```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
num_elem  = [ 1000,   10000, 30000, 50000, 100000, 200000, 500000, 1000000]
time_set  = [0.004,   0.041, 0.279, 0.817,  4.322, 27.215, 326.228, 1506.56]
time_tree = [0.002,   0.027, 0.097, 0.163,  0.325, 0.83, 1.983, 3.998]
```

```python
p1 = np.polyfit(num_elem, time_set, 3)
x = np.linspace(start=1000, stop=1000000, num=100000)
y_approx_set = np.polyval(p1, x)
p1
```

```python
p2 = np.polyfit(num_elem, time_tree, 3)

y_approx_tree = np.polyval(p2, x)
p2
```

```python

fig = plt.subplots(1, 2, figsize=(10, 5))

plt.subplot(121)
plt.scatter(num_elem, time_set, color='blue', label='std::set Points')
plt.plot(x, y_approx_set, color='red', label='std::set Approx')

plt.xlabel('Number of elements')
plt.ylabel('Time, s')
plt.title('std::set')

plt.grid(True)
plt.legend()

plt.subplot(122)

plt.scatter(num_elem, time_tree, color='black', label='AVL-tree Points')
plt.plot(x, y_approx_tree, color='green', label='AVL-tree Approx')

plt.xlabel('Number of elements')
plt.ylabel('Time, s')
plt.title('AVL-tree')

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.scatter(num_elem, time_set, color='blue', label='std::set Points')
plt.plot(x, y_approx_set, color='red', label='std::set Approx')

plt.xlabel('Number of elements')
plt.ylabel('Time, s')
plt.title('Compare between AVL-tree and std::set')

plt.scatter(num_elem, time_tree, color='black', label='AVL-tree Points')
plt.plot(x, y_approx_tree, color='green', label='AVL-tree Approx')

plt.grid(True)
plt.legend()
plt.show()
```
